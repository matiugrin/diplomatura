### Práctico 1

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
children = pd.read_csv('https://raw.githubusercontent.com/DiploDatos/AnalisisYCuracion/master/input/wtageinf.csv')
```

```
children.head()
```

```
boys = children[children.Sex == 1]
girls = children[children.Sex == 2]
```

```
sns.set(rc={'figure.figsize':(11.7,8.27)})
columns = ['P3','P5','P10','P25','P50','P75','P90','P95','P97']
for eje_y in columns:
  sns.lineplot(data=girls, x='Agemos', y=eje_y, label=eje_y)
plt.ylabel('Percentiles', fontsize=16)
plt.legend()
sns.despine()
```

```
sns.set(rc={'figure.figsize':(11.7,8.27)})
columns = ['P3','P5','P10','P25','P50','P75','P90','P95','P97']
for eje_y in columns:
  sns.lineplot(data=boys, x='Agemos', y=eje_y, label=eje_y)
plt.ylabel('Percentiles', fontsize=16)
plt.legend()
sns.despine()
```
