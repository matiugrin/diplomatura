```
import pandas as pd
import seaborn 
import matplotlib.pyplot as plt
```

```
dataset = pd.read_json('https://raw.githubusercontent.com/DiploDatos/AnalisisYCuracion/master/input/pasos.json')
```

```
dataset
```

```
provincias = dataset.groupby(by='province').count().reset_index()
provincias
```

```
seaborn.barplot(data=provincias, x='province', y='name', color='lightblue')
plt.xticks(rotation=90)
seaborn.despine()
```

```
seaborn.countplot(x=dataset.province.sort_values())
plt.xticks(rotation=90)
seaborn.despine()
```
